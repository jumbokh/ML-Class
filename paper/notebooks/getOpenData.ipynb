```python
import requests

r = requests.get("https://data.coa.gov.tw/Service/OpenData/FromM/FarmTransData.aspx", verify=False)
print(r.status_code)
```

```python
list_of_dicts = r.json()
print(type(r))
print(type(list_of_dicts))
#for i in list_of_dicts:
#    print(i["County"], i["SiteName"], i["PM2.5"])
```

```python
print(list_of_dicts)
```

```python
a_key = '作物名稱'
values_of_akey = [a_dict[a_key] for a_dict in list_of_dicts]
b_key = '作物代號'
values_of_bkey = [a_dict[b_key] for a_dict in list_of_dicts]
print(values_of_akey)
print(values_of_bkey)
```

```python
aList  = list(zip(values_of_akey,values_of_bkey))
print(aList)
```

```python
NameList = []
for a,b, in aList:
  NameList.append(str(a)+'-'+b)
print(NameList)
```

```python
#for i in range(len(NameList)):
print(NameList[:5])
```

```python
subs = '花椰菜'
res = [i for i in NameList if subs in i]
print(res)
```

### 處理日期

```python
c_key = '交易日期'
values_of_ckey = [a_dict[c_key] for a_dict in list_of_dicts]
d_key = '平均價'
values_of_dkey = [a_dict[d_key] for a_dict in list_of_dicts]
e_key =  '市場名稱'
values_of_ekey = [a_dict[e_key] for a_dict in list_of_dicts]
print(values_of_ckey)
print(values_of_dkey)
print(values_of_ekey)
```

```python
YearList = []
MonthList = []
DayList = []
YearMonth = []
for x in values_of_ckey:
  s = x.split('.')
#  print(s)
  Y = str(int(s[0])+1911)
  YearList.append(Y)
  MonthList.append(s[1])
  DayList.append(s[2])
  YearMonth.append(Y+'-'+s[1])
print(YearList)
print(MonthList)
print(DayList)
print(YearMonth)
```

```python
import pandas as pd
Transdf = pd.DataFrame(
    {'Year': YearList,
     'Month': MonthList,
     'Day': DayList,
     'YearMonth': YearMonth,
     'ItemNameID': NameList,
     'Market_name': values_of_ekey,
     'AveragePrice': values_of_dkey}
)
print(Transdf)
Transdf_old = Transdf.copy()
```

```python
#print(ndf.loc[ndf['AveragePrice']>0])
Taipeids = Transdf_old.loc[Transdf_old['Market_name']=='台北市場']
#print(Transdf_old.loc[Transdf_old['Market_name']=='台北市場'])
```

```python
print(Taipeids)
```

```python
Taipeids = Taipeids.loc[Taipeids['AveragePrice']>0.0]
print(Taipeids)
```

```python
Weekday = []
import calendar
for ro in Transdf.iterrows():
  #print(type(ro[1]))  # panadas DataFram Serials
  y = int(ro[1][0])
  m = int(ro[1][1])
  d = int(ro[1][2])
  w = calendar.weekday(y,m,d)
  Weekday.append(w)
print(Weekday)
```

```python
Transdf['Weekday'] = Weekday
print(Transdf.columns)
```
