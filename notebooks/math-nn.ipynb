```python
import numpy as np
import tensorflow as tf
import keras as keras
from keras import losses
```

```python
n = list(range(100))
y=[]
for x in n:
  i = (x*3)%100
  y.append(i)
print(y)
```

```python
import pandas as pd
dic = {
    "num": pd.Series(range(100)),
    "anti": y
}
```

```python
df = pd.DataFrame(dic)
df
```

```python
model = keras.Sequential()
model.add(keras.layers.Dense(units=256, input_dim=1, kernel_initializer='normal', activation='relu'))
model.add(keras.layers.Dense(units=100, kernel_initializer='normal', activation='softmax'))
model.compile(loss='mean_squared_error',optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
x = list(df['num'])
y = list(df['anti'])
```

```python
history = model.fit(x,y,batch_size=10,epochs=500)
```
