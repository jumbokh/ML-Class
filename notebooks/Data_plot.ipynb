```python
import numpy as np
import matplotlib.pyplot as plt
B = 100
N = 10000
np.random.seed(2)
x = np.random.normal(loc = 0, scale = 1, size = N)
y = np.random.standard_t(10000, size = N)
x_count, x_bin = np.histogram(x, bins = B)
y_count, y_bin = np.histogram(y, bins = B)
x_cumsum = np.cumsum(x_count)
y_cumsum = np.cumsum(y_count)
plt.figure(4)
plt.plot(list(range(B)), x_cumsum)
plt.plot(list(range(B)), y_cumsum)
plt.show()
plt.close()
```
