```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from som import SOM
```

```python
df = pd.read_csv('features.csv',encoding='unicode-escape')
```

```python
df[1:]
```

```python
cusid = df['Type']
```

```python
df = df.drop('Type', axis=1)
```

```python
df.shape
```

```python
som = SOM(x=10, y=10)
data = df.to_numpy()
som.fit(data, 10000, save_e=True, interval=100)
print("Fit error: %.4f" % som.error)
som.plot_error_history(filename="images/som_error.png")
```

```python
som.plot_distance_map(filename='images/distance_map.png')  # plot the distance map after training
```

```python
som.plot_density_map(data, filename="images/density.png")
```

```python
targets = np.array(8 * [0] + 8 * [1])  # create some dummy target values
```

```python
# now visualize the learned representation with the class labels
som.plot_point_map(data, targets, ['Class 0', 'Class 1'], filename='images/som.png')
som.plot_class_density(data, targets, t=0, name='Class 0', filename='images/class_0.png')
som.plot_distance_map(filename='images/distance_map.png')  # plot the distance map after training
```
