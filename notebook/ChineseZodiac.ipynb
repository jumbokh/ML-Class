```
def transform(year):
  head1 = "甲乙丙丁戊己庚辛壬癸"
  head2 = "子丑寅卯辰巳午未申酉戌亥"
  head3 = "癸己甲乙戊丙丁己庚辛戊壬"
  if year < 1930:
    print('請輸入西洋年1930以後')
    return -1
  else:
    n1 = (year-1930+6)%10
    n2 = (year-1930+6)%12
    msg = head1[n1]
    msg += head2[n2]
    return msg
```

```
def getn(n,x):
  ind='甲乙丙丁戊己庚辛壬癸'
  ind1='子丑寅卯辰巳午未申酉戌亥'
  n1 = x[0]
  n2 = x[1]
  o1 = tens[n]
  o2 = glist[n]
  i = ind.index(n1)
  r1 = o1[i]
  j = ind1.index(n2)
  r2 = o2[j]
  return (r1,r2)
```

```
def listname(inp1):
  base = inp1[0]
  sex = inp1[1]
  temp = inp1.copy()
  temp.remove(base)
  temp.remove(sex)
  fn = [['','','',''],
    ['','','',''],
    ['','','',''],
    ['','','','']]
  p=3
  for x in temp:
    (n1,n2)=getn(base,x)
    fn[0][p]=n1
    fn[1][p]=x[0]
    fn[2][p]=x[1]
    fn[3][p]=n2
    p=p-1
  fn[0][1]=sex
  for i in range(0,4):
    for j in range(0,4):
      print('%5s'%fn[i][j],end='')
    print()
```

```
def transm(D,month): # 立春: 2月, 庚, 寅, 2~12,1
  start=2
  if month<2:
    month=13
  head1 = "甲乙丙丁戊己庚辛壬癸"
  head2 = "子丑寅卯辰巳午未申酉戌亥"
  head3 = "癸己甲乙戊丙丁己庚辛戊壬"
  if month>13:
    print('請輸入1~12')
    return -1
  else:
    n1 = (month-start+6)%10
    n2 = (month-start+2)%12
    msg = head1[n1]
    msg += head2[n2]
    msg2=getn(D,msg)
    return msg,msg2
```

```
for i in range(2,13):
  print("%3d %s"%(i,transm('丙',i)))
print("%3d %s"%(1,transm('丙',1)))
```

```
# Python3 code to convert tuple  
# into string 
def convertTuple(tup): 
    str =  ''.join(tup) 
    return str
```
